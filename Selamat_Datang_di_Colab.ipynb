List berisi nama 5 mahasiswa

```
mahasiswa=["didi","dudu","dede","dodo","dada"]
print("list nama mahasiswa")
print(mahasiswa)
```

tambahakan 1 nama baru menggunakan append

```
mahasiswa.append("edi")
print(mahasiswa)
```

hapus 1 elemen menggunakan remove

```
mahasiswa.remove("didi")
print(mahasiswa)
```
